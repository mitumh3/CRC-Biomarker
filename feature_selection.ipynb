# 2. Feature Selection

```python
from process_func import select_features, model_name

# Select features based on ANOVA f-test
features = select_features(
    data,
    200,
    0.5,
    "Standard"
)
```

```python
(
    "224963_at" in features
    and "227331_at" in features
    and "230085_at" in features
    and "212514_x_at" in features
)
```

```python
import itertools

# Combine selected features to create model names -> 200 features -> more than 1M models will have to be built
print("\nMAKING COMBINATION...\n")
combinations_all = []
for i in range(1, 3+1):
    combinations_all.extend(list(itertools.combinations(features, i)))
features = combinations_all
print("\nADD FEATURES\n")
featurelist = []
for i in tqdm(features):
    selected_features = []
    for k in i:
        selected_features.append(k)
    feature_name = model_name(selected_features)
    featurelist.append(feature_name)
```

```python
with open("featurelist.csv", "w") as file:
    file.write(",".join(featurelist))
```
